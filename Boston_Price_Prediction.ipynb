```python
# Let's create a function to create adjusted R-Squared
def adj_r2(x,y):
    r2 = regression.score(x,y)
    n = df.shape[0]
    p = df.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
#creating any random features
random.seed(123)
nums = random.rand(df.shape[0])
mask_small = nums<.5
df['dummy'] = 'large'
df.loc[mask_small, 'dummy'] = 'small'
```

```python
from sklearn.linear_model import *
from sklearn.preprocessing import *
from sklearn.model_selection import *
```

```python
x_train,x_test,y_train,y_test = train_test_split(X , y, test_size = 0.25,random_state=355)
```

```python

lasscv = LassoCV(alphas = None,cv =10, max_iter = 100000, normalize = True)
lasscv.fit(x_train, y_train)
alpha = lasscv.alpha_
alpha
```

```python
lasso_reg = Lasso(alpha)
lasso_reg.fit(x_train, y_train)
lasso_reg.score(x_test, y_test)
```

```python
lassocv = LassoCV(alphas=None, cv=10,  max_iter=1000, normalize=True)
lassocv.fit(x_train, y_train)
```

```python
alpha = lassocv.alpha_
alpha
```

```python
lasso_reg = Lasso(alpha)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_reg.score(x_test, y_test)
```

```python
alphas = np.random.uniform(low=0, high=10, size=(50,))
ridgecv = RidgeCV(alphas = alphas,cv=10,normalize = True)
ridgecv.fit(x_train, y_train)
```

```python
ridgecv.alpha_
```

```python
ridge_model = Ridge(alpha=ridgecv.alpha_)
ridge_model.fit(x_train, y_train)
ridge_model.score(x_test, y_test)
```

```python
#elastic net
elastic_cv = ElasticNetCV(alphas=None, cv=10)
elastic_cv.fit(x_train, y_train)
elastic_cv.alphas_
```

```python
elasticnet = ElasticNet(alpha = elastic_cv.alpha_, l1_ratio=elastic_cv.l1_ratio)
elasticnet.fit(x_train, y_train)
elasticnet.score(x_test, y_test)
```
